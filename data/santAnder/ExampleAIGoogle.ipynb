```python
import pandas as pd 

df = pd.read_csv('train.csv',sep=',')
df.columns
```

```python
df.drop(columns=['ID'],inplace=True)
listCols = list(df.columns)
```

```python
listCols.remove('TARGET')
listReord = ['TARGET']
listReord.extend(listCols)
listReord
```

```python
df = df[listReord]
```

```python
df.to_csv('trainReord.csv',sep=',',header=False,index=False)
```
